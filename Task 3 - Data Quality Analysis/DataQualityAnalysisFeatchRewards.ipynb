## Data Quality Analysis

### Missing Values 

```python
def percentage_missing_values(series):
    df = pd.DataFrame(series.isna().sum()/series.shape[0], columns = ['% Missing  Values'])
    df.index.name = 'Column'
    return df
```

### Users

```python
percentage_missing_values(users)
```

```python
def analyze_relation(df,list_features):
    for column in list_features:
        df_temp = df[df[column].isnull()].drop(column,axis=1)
        print('------------------------')
        print('Column Analyzed ', column)
        for c in df_temp.columns:
            print('                          ')
            print('column', c)
            print('                          ')
            print(df_temp[c].value_counts())

insterest_features = ['lastLogin','signUpSource','state']            
analyze_relation(users,insterest_features)
```

The missing values in this dataset might  come from an integration error or an error in the systems. Because analizing the data, there is not a clear trend, except for the columns state and signUpSource. The missing data on those come from  users whose rol is **fetch-staff**

#### Brands

```python
percentage_missing_values(brands)
```

```python
interest_features = ['category','categoryCode','topBrand','brandCode']
analyze_relation(brands,interest_features)
```

Just like in the previous dataset, It seems like the missing values come from an error in the systems or in the data integration process. In this case It was not possible to identify a relationship between the columns with missing values and the rest of the columns.

##### Column rewardsReceiptItemList

The values in the column rewardsReceiptItemList present a lot of inconsistences. Each product in the list should have all the keys listed below. However, it seems like those keys are assigned dinamically per item. Most of the items only have the next keys:
    
    - partnerItemId
    - finalPrice
    - itemPrice
    - quantityPurchased
    - description
    - discountedItemPrice

```python
accesed = {}
for lis in  receipts.rewardsReceiptItemList:
   
    try:
        for dic in lis:
            #print(dic['brandCode'])
              
            for key in dic:
                
                if key not in accesed:
                    accesed[key] = 0
                else:
                    accesed[key] += 1
    except Exception as e:
        pass
dict(sorted(accesed.items(), key = lambda x:x[1], reverse= True))
```

### Duplicated rows 

#### Users Dataset

```python
users._id = [dit['$oid'] for dit in users._id]
users.createdDate = [dit['$date'] for dit in users.createdDate ]
users.drop('lastLogin',axis=1, inplace =True)
```

```python
user_id_count = users._id.value_counts()
user_id_count = user_id_count[user_id_count> 2]
user_id_count
```

```python
for id in user_id_count.index:
    filtered_id = users[users._id == id]
    
    print('-----------------------------------------------')
    print('user ', id)
    display(filtered_id[filtered_id.duplicated()])
    display(filtered_id[~filtered_id.duplicated()].iloc[0:1])
```

This dataset has a considerable amount of duplicated rows. In this case I found out tha there are users who are registered in the database up to 20 times, it means that there are 20 rows with exactly the same information. This maybe cause by the lack of validation of uniqueness in the user id.
